```python
from transformers import BertTokenizer
tokkenizer = BertTokenizer.from_pretrained('bert-base-uncased')
```

```python
from transformers import pipeline
```

```python
clasificador = pipeline('sentiment-analysis', model='dccuchile/bert-base-spanish-wwm-cased', tokenizer='dccuchile/bert-base-spanish-wwm-cased')
```

```python
textos = [
    "Me encto este producto, supera todas mis espectativas",
    "El servicio al cliente fue pesimo y el envio tardo demasido",
    "La calidad es decente, pero no es nada del otro mundo",
    "Era justo lo que buscaba",
    "Todos deberiamos tener uno"
]
```

```python
# Realizar la prediccion
prediccion = clasificador(textos)
```

```python
for texto, resultado in zip(textos, prediccion):
    sentimiento = resultado['label']
    puntuacion = resultados['score']
    print('-'*30)
    print(f'Texto: **{texto}**')
    print(f'Sentimiento: **{sentimiento}**')
    print(f'Confianza (score): **{puntuacion:.4f}**')
```
