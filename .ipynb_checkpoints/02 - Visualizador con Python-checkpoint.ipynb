```python
import pandas as pd
from matplotlib import pyplot as plt
import sklearn.datasets as ds
```

```python
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, engine="python", header=None)
print(raw_df)
```

```python
# Paso 1: las filas pares tienen 10 columnas, las impares tienen 3
data_part1 = raw_df.iloc[::2].reset_index(drop=True)  # columnas 0–9
data_part2 = raw_df.iloc[1::2, :3].reset_index(drop=True)  # columnas 10–12

# Paso 2: unir horizontalmente
boston_df = pd.concat([data_part1, data_part2], axis=1)

# Paso 3: asignar nombres de columnas
boston_df.columns = [
    "CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE", "DIS",
    "RAD", "TAX", "PTRATIO", "B", "LSTAT", "MEDV"
]
```

```python
print(boston_df.head())
```

```python
boston_df.plot(x='CRIM', y='MEDV', kind='scatter')
plt.title('Indice de criminalidad en un eje normal')
plt.show()
```

```python
boston_df.plot(x='CRIM', y='MEDV', kind='scatter', logx=True)
plt.title('Indice de criminalidad en un eje logaritmico')
plt.show()
```
