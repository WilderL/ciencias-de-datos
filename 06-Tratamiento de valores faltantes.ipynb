```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
datos = sns.load_dataset('titanic')
datos
```

```python
datos.info()
```

```python
datos.columns
```

```python
datos.drop('embarked', axis=1, inplace=True)
datos.drop('class', axis=1, inplace=True)
datos.drop('who', axis=1, inplace=True)
datos.drop('deck', axis=1, inplace=True)
datos.drop('embark_town', axis=1, inplace=True)
datos.drop('alive', axis=1, inplace=True)
datos.drop('alone', axis=1, inplace=True)
datos
```

```python
from sklearn.preprocessing import LabelEncoder
codificador = LabelEncoder()
datos['sex'] = codificador.fit_transform(datos['sex'])
```

```python
copiaDatos = datos.copy()
```

```python
# dividir el conjunto en X (explicativas) y Y (respuesta)
y = datos['survived']
datos.drop('survived', axis=1, inplace=True)
datos
```

```python
datos.info()
```

```python
print(datos.isnull().sum())
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(datos, y, test_size = 0.3, random_state=17)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(X_train, y_train)
```

## Eliminar columnas con valores faltantes

```python
datosActualizados = datos.dropna(axis=1)
datosActualizados
```

```python
X_train, X_test, y_train, y_test = train_test_split(datosActualizados, y, test_size=0.3, random_state=17)
lr.fit(X_train, y_train)
pred = lr.predict(X_test)
```

```python
pred
```

```python
from sklearn import metrics
```

```python
print(metrics.accuracy_score(pred, y_test))
```

## Rellenar datos faltante -inputar

imputar con:
 - **Media**: Promedio de las datos
 - **Mediana**: Valor central
 - **Moda**: El mas repetido (unimodal, bimodal, multimodal o no hay moda)

```python
datosRelleno = copiaDatos.copy()
datosRelleno['age'] = datosRelleno['age'].fillna(datosRelleno['age'].mean())
datosRelleno.info()
```

```python
# Segmentar los grupo en entrenamiento y pruebas
y = datosRelleno['survived']
datosRelleno.drop('survived', axis=1, inplace=True)
X_train, X_test, Y_train, Y_test = train_test_split(datosRelleno, y, test_size=0.3, random_state=17)
lr.fit(X_train, Y_train)
pred = lr.predict(X_test)
print(metrics.accuracy_score(pred, Y_test))
```

# Otras formas de imputatacion
- Interpolacion lineal
- Rellenar con el dato de lasiguiente fila
- Agregar una columna

```python
imputados = copiaDatos.copy()
imputados['ageMissing'] = imputados['age'].isnull()
imputados.info()
```

```python
from sklearn.impute import SimpleImputer
```

```python
# Segmentar los grupo en entrenamiento y pruebas
y = imputados['survived']
imputados.drop('survived', axis=1, inplace=True)
```

```python
imputador = SimpleImputer(strategy='median')
nuevoDatos = imputador.fit_transform(imputados)
imputados.info()
```

```python
nuevoDatos
```

```python
columnas = imputados.columns
imputados = pd.DataFrame(nuevoDatos, columns=columnas)
imputados
```

```python
# Segmentar los grupo en entrenamiento y pruebas
X_train, X_test, Y_train, Y_test = train_test_split(imputados, y, test_size=0.3, random_state=17)
lr.fit(X_train, Y_train)
pred = lr.predict(X_test)
print(metrics.accuracy_score(pred, Y_test))
```
